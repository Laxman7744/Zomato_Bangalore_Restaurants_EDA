### Importing Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('dark_background')
```

### Reading CSV

```python
df = pd.read_csv("zomato.csv")
df.head()
```

```python
df.shape # rows and columns
```

```python
df.columns
```

```python
df = df.drop(['url','address','menu_item','phone','dish_liked','reviews_list'], axis = 1)
df.head()
```

```python
df.info() # 
```

### Dropping Duplicates

```python
df.drop_duplicates(inplace = True)
df.shape
```

### Cleaning Rate Column

```python
df['rate'].unique()
```

### Removing "NEW","-" and "/5" from Rate Column

```python
def handlerate(value):
    if (value == 'NEW' or value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

```python
df.rate.isnull().sum()
```

### Filling Null Values In Rate Column with Mean

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

### Dropping Null Values

```python
df.dropna(inplace = True)
df.head()
```

```python
# Rename the columns name
df.rename(columns = {'approx_cost(for two people)':'Cost2plates','listed_in(type)':'Type'},inplace=True)
```

```python
df.head()
```

```python
df['location'].unique()
```

```python
df['listed_in(city)'].unique()
```

### Listed in(city) and location,both are there, lets keep only one.

```python
### listed in(city) droped beacause of so many unique values
df = df.drop(['listed_in(city)'], axis = 1)
```

```python
df['Cost2plates'].unique()
```

### Removing , from Cost2plates Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
    return float(value)

df['Cost2plates'] = df['Cost2plates'].apply(handlecomma)
print(df['Cost2plates'].unique())
```

```python
df.tail()
```

### Cleaning Rest Type Column

```python
df['rest_type'].value_counts()
```

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types < 100000]
rest_types_lessthan1000
```

### Making Rest Types less than 1000 in frequency as others

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.tail(2)
```

### Cleaning location column

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending = False)

location_lessthan300 = location [location < 300]


def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

### Cleaning Cuisines Column

```python
cuisines = df['cuisines'].value_counts(ascending = False)

cuisines_lessthan300 = cuisines[cuisines < 300]


def handle_cuisines(value):
    if(value in cuisines_lessthan300):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

### Visualizing Online Order Facility, Locationwise

```python
# Group by location and online_order, then count restaurant names
df1 = df.groupby(['location', 'online_order'])['name'].count().reset_index()

# Save to CSV
df1.to_csv('location_online.csv', index=False)

# Read the saved CSV
df1 = pd.read_csv('location_online.csv')

# Create pivot table
df1 = pd.pivot_table(df1, 
                     values='name', 
                     index='location', 
                     columns='online_order', 
                     fill_value=0, 
                     aggfunc=np.sum)

df1
```
